```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("awaiskaggler/insurance-csv")

print("Path to dataset files:", path)
```

```python
import pandas as pd

df = pd.read_csv(path + "/insurance.csv")
df.head()
```

```python
#Create vectors
import numpy as np
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['sex'] = le.fit_transform(df['sex'])
df['smoker'] = le.fit_transform(df['smoker'])
df['region'] = le.fit_transform(df['region'])
df.head()
```

```python
#create factors
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
print(x)
print(y)
```

```python
#create lists
x_list = x.tolist()
y_list = y.tolist()
print(x_list)
print(y_list)
```

```python
#create matrices
x_matrix = np.array(x_list)
y_matrix = np.array(y_list)
print(x_matrix)
print(y_matrix)

```

```python
#create data frames
x_df = pd.DataFrame(x_matrix)
y_df = pd.DataFrame(y_matrix)
print(x_df)
print(y_df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Filter data for age between 20 and 40
filtered_df = df[(df['age'] >= 20) & (df['age'] <= 40)]

# Group by age and smoking status, then count occurrences
age_smoking_counts = filtered_df.groupby(['age', 'smoker'])['smoker'].count().reset_index(name='count')

# Get unique age values
age_values = age_smoking_counts['age'].unique()

# Create lists for smoker and non-smoker counts
smoker_counts = []
non_smoker_counts = []

# Iterate through age values and get counts for each smoking status
for age in age_values:
    smoker_counts.append(age_smoking_counts[(age_smoking_counts['age'] == age) & (age_smoking_counts['smoker'] == 1)]['count'].values[0] if len(age_smoking_counts[(age_smoking_counts['age'] == age) & (age_smoking_counts['smoker'] == 1)]) > 0 else 0)  # Handle cases where an age group has no smokers
    non_smoker_counts.append(age_smoking_counts[(age_smoking_counts['age'] == age) & (age_smoking_counts['smoker'] == 0)]['count'].values[0] if len(age_smoking_counts[(age_smoking_counts['age'] == age) & (age_smoking_counts['smoker'] == 0)]) > 0 else 0)  # Handle cases where an age group has no non-smokers

# Create bar chart
width = 0.35  # Width of the bars
x = np.arange(len(age_values))  # X-axis positions for the bars

plt.bar(x - width/2, smoker_counts, width, label='Smoker', color='red')
plt.bar(x + width/2, non_smoker_counts, width, label='Non-Smoker', color='blue')

# Customize the plot
plt.title('Bar Chart of Age (20-40) vs. Smoking')
plt.xlabel('Age')
plt.ylabel('Count')
plt.xticks(x, age_values)  # Set x-axis ticks and labels
plt.legend()  # Show legend

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate average expenses for smokers and non-smokers
smoker_expenses = df[df['smoker'] == 1]['expenses'].mean()
non_smoker_expenses = df[df['smoker'] == 0]['expenses'].mean()

# Create bar chart
labels = ['Smoker', 'Non-Smoker']
expenses = [smoker_expenses, non_smoker_expenses]

plt.bar(labels, expenses, color=['red', 'blue'])

plt.title('Average Expenses of Smokers vs. Non-Smokers')
plt.xlabel('Smoking Status')
plt.ylabel('Average Expenses')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Group data by sex and smoker, then count occurrences
smoker_gender_counts = df.groupby(['sex', 'smoker'])['smoker'].count().reset_index(name='count')

# Get smoker counts for each gender
female_smoker_count = smoker_gender_counts[(smoker_gender_counts['sex'] == 0) & (smoker_gender_counts['smoker'] == 1)]['count'].values[0]
male_smoker_count = smoker_gender_counts[(smoker_gender_counts['sex'] == 1) & (smoker_gender_counts['smoker'] == 1)]['count'].values[0]

# Create bar chart
labels = ['Female', 'Male']
smoker_counts = [female_smoker_count, male_smoker_count]

plt.bar(labels, smoker_counts, color=['pink', 'lightblue'])

plt.title('Number of Smokers by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Smokers')

plt.show()
```
